```python
df4 = pd.read_csv('df4.csv')
df4 = df4.drop(['Unnamed: 0'], axis = 1)
```

```python
#Function to select only state from the work_location column
def select_state(location):
    state = location.split(',')[1].strip().lower()
    return state


df4['state'] = df4['work_location'].apply(select_state)
```

```python
#Grouping location based on common states
(df4.groupby('state')['case_status'].count()
.reset_index()
.sort_values('case_status', ascending = False).head())
df4.head()
```

```python
#list of states by region
west = ['alaska','arizona', 'california', 'colorado', 'hawaii', 
            'idaho', 'montana', 'nevada', 'new nexico', 'oregon', 'utah', 'washington', 'wyoming']
mid_west = ['north dakota', 'south dakota', 'nebraska', 
              'minnesota', 'iowa', 'missouri', 'wisconsin', 
              'illinois', 'kansas', 'michigan', 'indiana', 'ohio']
north_east = ['connecticut', 'delaware', 'maine', 'massachusetts', 
              'maryland', 'new hampshire', 'new jersey', 'new york', 
              'pennsylvania', 'rhode island', 'vermont']
south = ['arkansas', 'louisiana', 'oklahoma', 'texas', 'georgia', 'maryland',
             'texus', 'north carolina', 'south carolina']
```

```python
#Make new columns with region names and fill the states with region
df4['west'] = df4['state'].isin(west)
df4['west'] = df4.apply((lambda x: 'west' if x.west else 'NAN'), axis = 1)
df4['mid_west'] = df4['state'].isin(mid_west)
df4['mid_west'] = df4.apply((lambda x: 'mid_west' if x.mid_west else 'NAN'), axis = 1)
df4['north_east'] = df4['state'].isin(north_east)
df4['north_east'] = df4.apply((lambda x: 'north_east' if x.north_east else 'NAN'), axis = 1)
df4['south'] = df4['state'].isin(south)
df4['south'] = df4.apply((lambda x: 'south' if x.south else 'NAN'), axis = 1)
```

```python
df4['north_east'].replace(['north_east', 'NAN'], [1,0], inplace = True)
df4['west'].replace(['west', 'NAN'], [1,0], inplace = True)
df4['mid_west'].replace(['mid_west', 'NAN'], [1,0], inplace = True)
df4['south'].replace(['south', 'NAN'], [1,0], inplace = True)
```
